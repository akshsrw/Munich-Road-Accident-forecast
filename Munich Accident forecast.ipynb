```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
accident_data= pd.read_csv('210619.csv')
accident_data
```

```python
## Remove extra columns
accident_data.drop(['VORJAHRESWERT','VERAEND_VORMONAT_PROZENT','VERAEND_VORJAHRESMONAT_PROZENT','ZWOELF_MONATE_MITTELWERT'], axis=1, inplace=True)
accident_data.head(20)
```

```python
##Drop empty rows and 'Summe' row from MONAT 

accident_data.dropna(subset=['WERT'],inplace=True)
accident_data=accident_data[accident_data.MONAT.apply(lambda x: x.isnumeric())]
accident_data.head(20)
```

```python
##Drop first four characters from MONAT column

accident_data['MONAT']=accident_data['MONAT'].str[4:]
accident_data.head()
```

```python
##Rename year and month 

accident_data=accident_data.rename(columns={'JAHR':'year', 'MONAT':'month'})
accident_data
```

```python
##Create a DATE column

accident_data['DATE'] = pd.to_datetime(accident_data[['year', 'month',]].assign(DAY=1))

accident_data.set_index('DATE',inplace=True)
accident_data
```

```python
accident_data['MONATSZAHL'].unique()  ##CHECK all the unique values in MONATSZAHL
```

```python
## PLOT ALL VALUES INDIVIDUALLY

p=accident_data ## WORKING ON COPY OF DATASET
 
figure, axis = plt.subplots(3, 2,figsize=(15,10)) 


for ci, i in enumerate(['Alkoholunfälle','Fluchtunfälle','Verkehrsunfälle']):
    for cj, j in enumerate(['insgesamt','Verletzte und Getötete']):
        P=accident_data[(accident_data['MONATSZAHL']==i) & (accident_data['AUSPRAEGUNG']==j)]
        P.sort_values('DATE', inplace=True)
        axis[ci, cj].plot(P['WERT'],)
        axis[ci, cj].set_title('%s -- %s!' % (i, j))
plt.show()
```

```python
##Seperate Training Dataset from the table

Dataset=accident_data[(accident_data['MONATSZAHL']=='Alkoholunfälle') & (accident_data['AUSPRAEGUNG']=='insgesamt')]
Dataset=Dataset[['WERT']]
Dataset.sort_values('DATE', inplace=True)
Dataset.head()
```

```python
## Checking the trend, seasonal behavior and Noise of the Dataset


from statsmodels.tsa.seasonal import seasonal_decompose 

seasonal = seasonal_decompose(Dataset,model='addition',period=12)
fig = plt.figure()  
fig = seasonal.plot()  
fig.set_size_inches(15, 10)
```

```python
## The model has Seasonal behavour, First Decreasing then Increasing Trend and high coarseness of noise.
## Checking if the data is stationary or not with AdFuller test (although the graphs shows its not stationary)

from statsmodels.tsa.stattools import adfuller

test_result=adfuller(Dataset['WERT'])

def adfuller_test(sales):
    result=adfuller(sales)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
adfuller_test(Dataset['WERT'])
```

```python
## Since the dataset has seasonality we will remove the seasonality and again check for stability

Dataset['Seasonal First Difference']=Dataset['WERT']-Dataset['WERT'].shift(12)
Dataset.head(20)
```

```python
adfuller_test(Dataset['Seasonal First Difference'].dropna())

```

```python
## Finding the Parameters of the model using PACF and ACF model

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig = plt.figure(figsize=(15,10))
ax1 = fig.add_subplot(211)
fig = plot_acf(Dataset['Seasonal First Difference'].iloc[13:],lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = plot_pacf(Dataset['Seasonal First Difference'].iloc[13:],lags=40,ax=ax2)
```

```python
# Training the Sarima model with pdq parameters as 2,2,2 and seasonal order 1,1,1,12

import statsmodels.api as sm
model=sm.tsa.statespace.SARIMAX(Dataset['WERT'],order=(2, 2, 2),seasonal_order=(1,1,1,12))
result=model.fit()
```

```python
Dataset['forecast']=result.predict(start=228,end=252,dynamic=True)
Dataset[['WERT','forecast']].plot(figsize=(16,8))
```
